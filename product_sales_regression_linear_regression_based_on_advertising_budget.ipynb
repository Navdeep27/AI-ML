step 1: import the libraries

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error
```

step 2: reading the data and splitting it to input and output

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/Navdeep27/AI-ML/refs/heads/main/product_sales_regression_based_on_advertising_budget_data.csv')
print(dataset)
inputx = dataset.iloc[:, :-1].values
outputy = dataset.iloc[:, 1].values
print(inputx)
print(outputy)
```

step 3: select 30 percent of the data for testing and 70 percent for training

```python
input_train, input_test, output_train, output_test = train_test_split(inputx, outputy, test_size = 0.3, random_state = 43)
print(input_test)
```

step 4: Creating a simple Linear Regression model

```python
model = LinearRegression()
print("\nThe parameters of the model are\n\n",model.get_params())
```

Step 5: Train the model

```python
print("\nThe model we are using is ", model.fit(input_train, output_train))
```

step 6: Printing the testing results

```python
print("\nThe test input (TV Budget ($),Radio Budget ($),Newspaper Budget ($)) is as follows \n")
print(input_test)
# model predicting the Test set results
predicted_output = model.predict(input_test)
print("\nThe output Sales (units) for the test input is as follows \n")
print(predicted_output)
```

step 7: Visualising the training results

```python
plt.figure(figsize=(12, 6))

# Plotting TV Budget vs Sales
plt.subplot(1, 3, 1)
# Accessing the first column (TV Budget) using integer indexing (0)
plt.scatter(input_train[:, 0], output_train, color='blue', label='Training data')
plt.plot(input_train[:, 0], model.predict(input_train), color='red', label='Regression model')
plt.xlabel('TV Budget ($)')
plt.ylabel('Sales (units)')
plt.title('TV Budget vs Sales')
plt.legend()

# Plotting Radio Budget vs Sales
plt.subplot(1, 3, 2)
# Accessing the second column (Radio Budget) using integer indexing (1)
plt.scatter(input_train[:, 1], output_train, color='blue', label='Training data')
plt.plot(input_train[:, 1], model.predict(input_train), color='red', label='Regression model')
plt.xlabel('Radio Budget ($)')
plt.ylabel('Sales (units)')
plt.title('Radio Budget vs Sales')
plt.legend()

# Plotting Newspaper Budget vs Sales
plt.subplot(1, 3, 3)
# Accessing the third column (Newspaper Budget) using integer indexing (2)
plt.scatter(input_train[:, 2], output_train, color='blue', label='Training data')
plt.plot(input_train[:, 2], model.predict(input_train), color='red', label='Regression model')
plt.xlabel('Newspaper Budget ($)')
plt.ylabel('Sales (units)')
plt.title('Newspaper Budget vs Sales')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

step 8: Printing the model efficiency

```python
# Calculate the mean squared error
mse = mean_squared_error(output_test, predicted_output)
rmse = root_mean_squared_error(output_test, predicted_output)
mae = mean_absolute_error(output_test, predicted_output)
print(mse)
print(rmse)
print(mae)
```

step 9: Testing the model prediction using testinput

```python
all_budgets = input("\nGive budget($) for TV, Radio, Newspaper) : ")
all_budgets = all_budgets.split(',')
budget = [float(i) for i in all_budgets]
testinput = [budget]
predicted_output = model.predict(testinput)
print('\nThe budget($) for TV, Radio, Newspaper is ',testinput)
print('\nThe sales for the budget($) for TV, Radio, Newspaper is ',predicted_output)
```
